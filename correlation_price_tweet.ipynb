## Load base data price action and twitter feed

```python
list_pa_files=os.listdir("./price_action_data//")
list_tweet_files=os.listdir("./data_tweet/")
```

```python
def rounder_price_fetch(t,pa_df):
    upper_bound_t=pd.to_datetime(t).ceil(freq='H')
    lower_bound_t=pd.to_datetime(t).floor(freq='H')
    price_diff=pa_df.loc[upper_bound_t]['close']-pa_df.loc[lower_bound_t]['close']
    return(price_diff)
```

```python
def fetch_pa_bytimeFrame(timeframe="1h"):
    ret_list=[]
    for fil in list_pa_files:
        if timeframe in fil:
            print(fil)
            ret_list.append(fil)
    return(ret_list)
```

```python
def fetch_tweetData():
    tweet_data_coll=pd.DataFrame()
    for fil in list_tweet_files:
        temp=pd.read_csv("./data_tweet/"+fil)
        tweet_data_coll=pd.concat([tweet_data_coll,temp])

    return(tweet_data_coll)
```
