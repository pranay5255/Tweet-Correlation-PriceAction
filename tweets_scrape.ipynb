```python
def twint_to_pandas(columns):
    return twint.output.panda.Tweets_df[columns]

def get_user_tweets(username,start_date='2022-01-01'):
    c = twint.Config()
    c.Username = username
#     c.Limit = 1
    c.Since = start_date
    c.Pandas = True

    # Run
    twint.run.Search(c)
    
    df_pd = twint.output.panda.Tweets_df
    
    return(df_pd)  

start_date='2022-01-01'
for user in list(all_users)[36:]:

    print(user)
    tweet_df=get_user_tweets(username=user.replace("@",""),start_date='2022-01-01')
    print(tweet_df.shape)
    tweet_df.to_csv("./data_tweet/"+user+"_"+start_date+"_.csv")
```
