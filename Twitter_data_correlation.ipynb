```python
def get_tweet_ticker(ticker,start_date='2022-01-01'):
    c = twint.Config()
    c.Search=[ticker]
    c.Limit = 1
    c.Since = start_date

    # Run
    twint.run.Search(c)
    Tweets_df = twint.storage.panda.Tweets_df
    return(tweets_df)
    
```
